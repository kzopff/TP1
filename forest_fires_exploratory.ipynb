```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('forestfires.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.dtypes
```

```python
month = df.month.value_counts()
month
```

```python
day = df.day.value_counts()
day
```

```python
df.corr()
```

```python
corr = df.corr()
```

```python
sns.heatmap(corr, annot=True)
plt.savefig('corr_heatmap.png')
```

```python
fig = plt.figure(figsize=(15,22))
ax1 = fig.add_subplot(6,2,1)
ax1.hist(df.X)
ax1.set_title('X')
ax2 = fig.add_subplot(6,2,2)
ax2.hist(df.Y)
ax2.set_title('Y')
ax3 = fig.add_subplot(6,2,3)
ax3.hist(df.FFMC)
ax3.set_title('FFMC')
ax4 = fig.add_subplot(6,2,4)
ax4.hist(df.DMC)
ax4.set_title('DMC')
ax5 = fig.add_subplot(6,2,5)
ax5.hist(df.DC)
ax5.set_title('DC')
ax6 = fig.add_subplot(6,2,6)
ax6.hist(df.ISI)
ax6.set_title('ISI')
ax7 = fig.add_subplot(6,2,7)
ax7.hist(df.temp)
ax7.set_title('Temp')
ax8 = fig.add_subplot(6,2,8)
ax8.hist(df.RH)
ax8.set_title('RH')
ax9 = fig.add_subplot(6,2,9)
ax9.hist(df.wind)
ax9.set_title('Wind')
ax10 = fig.add_subplot(6,2,10)
ax10.hist(df.rain)
ax10.set_title('Rain')
ax11 = fig.add_subplot(6,2,11)
ax11.hist(df.area)
ax11.set_title('Area')
plt.savefig('densityplts.png')
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, hue='month')
plt.savefig('pairplot.png')
```

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = month.sort_values().plot.barh()
ax1.set_title('Month')
ax2 = fig.add_subplot(1,2,2)
ax2 = day.sort_values().plot.barh(color='red')
ax2.set_title('Day')
plt.savefig('barcharts.png')
```
